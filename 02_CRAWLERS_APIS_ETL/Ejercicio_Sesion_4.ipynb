```python
nombre_fondo = ['BGF Euro Short Duration Bond A2 EUR',
'iShares Euro Govt Bd Idx (LU) A2 EUR',
'iShares Em Mkts Govt Bd Idx (LU) A2 EURH',
'iShares Europe Equity Index (LU) A2 EUR',
'iShares Dev Rl Ett Idx (IE) Instl Acc ‚Ç¨',
'iShares North America Eq Idx (LU) A2 USD',
'iShares Japan Equity Index (LU) A2 USD',
'iShares Emerging Mkts Eq Idx (LU) A2 USD']

participaciones = [22.45,
                   8.34,
                   13.22,
                   11.77,
                   31.32,
                   7.41,
                   6.22,
                   9.77]

urls = ['https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F0GBR04AS2',
       'https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F00000OYMY',
       'https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F00000X4QX',
       'https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F00000OYMI',
       'https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F00000PJME',
       'https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F00000OYNC',
       'https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F00000OYN8',
       'https://www.morningstar.es/es/funds/snapshot/snapshot.aspx?id=F00000P5RU']
```

```python
import urllib
from bs4 import BeautifulSoup
import numpy as np

# Conseguimos el tipo de cambio de dolar a euro
url_eu_dl = 'https://www.eleconomista.es/cruce/EURUSD'
html = urllib.request.urlopen(url_eu_dl)
soup = BeautifulSoup(html)
tags = soup.find_all('span', class_='ultimo_321 last-value')
cambio = float(tags[0].contents[0].replace('\xa0$/‚Ç¨','').replace(',','.'))

# Recorremos las urls y recogemos el valor
# cambiandolo a euros si el valor estuviera en dolars
valores = []

for url in urls:
    html = urllib.request.urlopen(url)
    soup = BeautifulSoup(html)
    tags = soup.find_all('td', class_='line text')
    
    moneda = tags[0].contents[0].split('\xa0')[0]
    if moneda == 'USD':
        valores.append(float(tags[0].contents[0].split('\xa0')[1].replace(',','.'))*cambio)
    else:
        valores.append(float(tags[0].contents[0].split('\xa0')[1].replace(',','.')))

valor_total = sum(np.array(valores) * np.array(participaciones))
print ('Valor total: ', valor_total, '‚Ç¨')
```

SE PIDE:
Implementar un crawler de la web http://www-math.mit.edu/~gs/ de dos niveles.
Para ello:
1) Extrae todos los hiperv√≠nculos de la web e impr√≠melos por pantalla

2) Accede a cada uno de ellos y vuelve a extraer los nuevos hiperv√≠nculos, imprimi√©ndolos tambi√©n por pantalla.

```python
# Extraemos todos los hipervinculos y los imprimimos
import urllib
from bs4 import BeautifulSoup

url = 'http://www-math.mit.edu/~gs/'
html = urllib.request.urlopen(url)

soup = BeautifulSoup(html)

tags = soup('a')

vinculos = []
for tag in tags:
#    print('TAG:', tag)
    print('URL:', tag.get('href'))
    vinculos.append(tag.get('href'))
#    print('Text:', tag.contents)
#    print('Attributes:', tag.attrs)
#    print('\n')
```

```python
# Accedemos a cada uno de ellos y los imprimimos en pantalla
for url in vinculos:
    try:
        html = urllib.request.urlopen(url)
        soup = BeautifulSoup(html)
        tags = soup('a')
        print ('URL: ', url)
        for tag in tags:
            print ('    SUB_URL:', tag.get('href'))
    except:
        print ('URL sin hipervinculos:', url)
```
