# Series temporales representaciÃ³n

```python
# !pip install seaborn
```

```python
import numpy as np
import pandas as pd

np.random.seed(0)
x = np.cumsum(np.random.normal(2,1,100))
y = np.cumsum(np.random.normal(0.5,1,100))

df = pd.DataFrame({'serie1':x, 'serie2': y})

df.head()
```

```python
import seaborn as sns
%matplotlib inline
# %matplotlib qt
```

```python
sns.tsplot(df.serie1,color='r');
sns.tsplot(df.serie2,color='b');
```
