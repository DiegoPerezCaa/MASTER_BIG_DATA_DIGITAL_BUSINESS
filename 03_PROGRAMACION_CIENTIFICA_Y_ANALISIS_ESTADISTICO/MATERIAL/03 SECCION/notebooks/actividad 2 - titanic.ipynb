6. Vamos a introducir medianas por categorías en la edad para suplir los valores faltantes basándonos en el género, la clase y el puerto de embarque. Primero introduciremos la moda en el puerto de embarque. Luego hacemos el agrupamiento calculando las medianas y generamos una función que nos permita hacer el rellenado por categorías.



```python
import seaborn as sns
%matplotlib inline

dat=sns.load_dataset('titanic')
print(dat.shape)
dat.head(10)
```

```python
dat.isnull().sum()/dat.shape[0]*100
```

```python
del dat['deck']
```

```python
dat.dtypes
```

```python
listacat=['pclass','sex','embarked','who','adult_male','alone']

for k in listacat:
    print(dat.groupby(k)['survived'].mean().sort_values(ascending=False))
    print('\n')
```

```python
dat.groupby(['sex', 'pclass'])[['age', 'fare']].mean()
```

Observamos que ser adulto masculino es muy condicionante para ser superviviente. También es favorable ir solo.

```python
len(listacat)
```

```python
import matplotlib.pyplot as plt
f,ax=plt.subplots(2,3,sharex=False,sharey=False);
f.subplots_adjust(wspace=0.4,hspace=0.6)
# f.tight_layout();
for i,k in enumerate(ax.ravel()):
    sns.factorplot(x=listacat[i],y='survived',data=dat,ax=k);
```

```python
f
```

Se concluyen de las gráficas:

- La probabilidad de supervivencia decrece conforme la clase del billete es más pobre.
- La probabilidad de supervivencia es mucho mayor para las mujeres.
- Los embarcados en el puerto C tuvieron más supervivencia.
- Ir acompañado favorecía la supervivencia.
- No ser un adulto masculino era un motivo muy fuerte de supervivencia.

Importamos de nuevo el dataset, vamos a meter la edad según la clase, el sexo y puerto de embarque. Primero rellenamos el valor del puerto de embarque con la moda.

```python
dat.isnull().sum()/dat.shape[0]
```

```python
dat.embarked.mode()
```

```python
#introducimos la moda en los puertos embarcados
dat.fillna({'embarked':dat.embarked.mode()},inplace=True);
```

Agrupamos por estas tres categorías:

```python
dat.groupby(['sex','class','embarked'])['age'].agg(['median','count'])
```

```python
dat['Age']=dat.groupby(['sex','class','embarked'])['age'].transform(lambda x: x.fillna(x.median()));
dat['Age'].isnull().sum()
```
