# Seaborn

Seaborn es una librería creada por Stanford para la representación estadística. Tiene estilos sofisticados e incorpora los tipos de gráficos esenciales para analizar distribuciones y relaciones estadísticas

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
#representación matplotlib 
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

Activamos el estilo Seaborn y pintamos lo mismo para observar la diferencia

```python
#representación Seaborn
import seaborn as sns
sns.set()
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

## Estilos Seaborn

En la función `sns.axes_style` podemos establecer los estilos

    Parameters
    ----------
    style : dict, None, or one of {darkgrid, whitegrid, dark, white, ticks}
        A dictionary of parameters or the name of a preconfigured set.
    rc : dict, optional
        Parameter mappings to override the values in the preset seaborn
        style dictionaries. This only updates parameters that are
        considered part of the style definition.

## Gráficos de distribución


- jointplot(x, y[, data, kind, stat_func, ...])	Draw a plot of two variables with bivariate and univariate graphs.
- pairplot(data[, hue, hue_order, palette, ...])	Plot pairwise relationships in a dataset.
- distplot(a[, bins, hist, kde, rug, fit, ...])	Flexibly plot a univariate distribution of observations.
- kdeplot(data[, data2, shade, vertical, ...])	Fit and plot a univariate or bivariate kernel density estimate.
- rugplot(a[, height, axis, ax])	Plot datapoints in an array as sticks on an axis.



```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

### Jointplot

Gráfico de dispersión con histogramas adyacentes entre dos variables

```python
with sns.axes_style('dark'):
    sns.jointplot('sepal_length','sepal_width',data=iris);
```

```python
with sns.axes_style('white'): 
    sns.jointplot('sepal_length','sepal_width',data=iris,kind='reg');
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg');
```

```python
with sns.axes_style('ticks'): 
    sns.jointplot('sepal_length','sepal_width',data=iris,kind='kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot("sepal_width", "sepal_length", iris, kind='hex')
```

### Pairplot

Hace gráficas de dispersión entre variables con diagonal en forma de histograma, sus parámetros:

        Parameters
        ----------
        data : DataFrame
            Tidy (long-form) dataframe where each column is a variable and
            each row is an observation.
        hue : string (variable name), optional
            Variable in ``data`` to map plot aspects to different colors.
        hue_order : list of strings
            Order for the levels of the hue variable in the palette
        palette : dict or seaborn color palette
            Set of colors for mapping the ``hue`` variable. If a dict, keys
            should be values  in the ``hue`` variable.
        vars : list of variable names, optional
            Variables within ``data`` to use, otherwise use every column with
            a numeric datatype.
        {x, y}_vars : lists of variable names, optional
            Variables within ``data`` to use separately for the rows and
            columns of the figure; i.e. to make a non-square plot.
        kind : {'scatter', 'reg'}, optional
            Kind of plot for the non-identity relationships.
        diag_kind : {'hist', 'kde'}, optional
            Kind of plot for the diagonal subplots.
        markers : single matplotlib marker code or list, optional
            Either the marker to use for all datapoints or a list of markers with
            a length the same as the number of levels in the hue variable so that
            differently colored points will also have different scatterplot
            markers.
        size : scalar, optional
            Height (in inches) of each facet.
        aspect : scalar, optional
            Aspect * size gives the width (in inches) of each facet.
        dropna : boolean, optional
            Drop missing values from the data before plotting.
        {plot, diag, grid}_kws : dicts, optional
            Dictionaries of keyword arguments.

```python
with sns.axes_style('darkgrid'): 
    sns.pairplot(iris);
```

Se puede establecer una columna con categorías para pintar con diferentes colores los puntos  e histogramas, además se pueden seleccionar las variables a trabajar del `DataFrame`

```python
sns.pairplot(iris,hue='species',vars=['sepal_length','sepal_width','petal_length']);
```

```python
iris.species.unique()
```

Introduciendo los valores de especies en `palette` y colores podemos seleccionar qué colores queremos, con `hue_order` se puede establecer el orden de los distintos niveles al pintar los histogramas. También se pueden cambiar histogramas por gráficas de densidad `kde` modificando el parámetro `diag_kind`.

```python
sns.pairplot(iris,hue='species',
             hue_order=['versicolor','virginica','setosa'],
             palette={'setosa':'r','versicolor':'b','virginica':'y'},
            diag_kind='kde');
```

Se pueden seleccionar variables para cada eje

```python
sns.pairplot(iris,x_vars=['sepal_length','petal_length'],y_vars=['sepal_width'],hue='species');
```

Añadir marcadores en los gráficos de dispersión para cada clase

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"]);
```

Hacer gráficas que contengan la recta de regresión en los scatterplots

```python
sns.pairplot(iris, kind="reg");
```

### Distplot

Gráfico de distribución de una variable

```python
sns.set(rc={"figure.figsize": (5,2)})
```

```python
sns.distplot(iris['sepal_length']);
```

```python
sns.distplot(iris.sepal_length, rug=True, hist=False);
sns.distplot(iris.petal_length, rug=True, hist=False);
```

```python
sns.distplot(iris.sepal_length, vertical=True);
```

### KDEplot

Representa una gráfica de densidad 1D ó 2D

```python
sns.kdeplot(iris.sepal_length);
```

El parámetro `bw` indica la anchura de análisis gaussiano aplicado:

```python
sns.kdeplot(iris.sepal_length, bw=.15);
```

```python
sns.kdeplot(iris.sepal_length,iris.petal_length);
```

```python
sns.kdeplot(iris.sepal_length,iris.petal_length,n_levels=30,cmap='viridis');
```

### Rugplot

Ticks en un eje, se suele combinar como opción en gráficos anteriores

```python
sns.rugplot(iris.sepal_length);
```
