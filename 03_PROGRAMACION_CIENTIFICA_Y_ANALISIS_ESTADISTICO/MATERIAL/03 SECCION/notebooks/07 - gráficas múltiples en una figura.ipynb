# Creando varias gráficas en la misma figura

Matplotlib ofrece configuraciones precisas para crear gráficas múltiples en una figura con la función `subplots`

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

      *sharey* : string or bool
        If *True*, the Y axis will be shared amongst all subplots. If
        *True* and you have multiple columns, the y tick labels on all but

        If a string must be one of "row", "col", "all", or "none".
        "all" has the same effect as *True*, "none" has the same effect
        as *False*.
        If "row", each subplot row will share a Y axis and the y tick

        If "col", each subplot column will share a Y axis.

      *squeeze* : bool
        If *True*, extra dimensions are squeezed out from the
        returned axis object:

        - if only one subplot is constructed (nrows=ncols=1), the
          resulting single Axis object is returned as a scalar.

        - for Nx1 or 1xN subplots, the returned object is a 1-d numpy
          object array of Axis objects are returned as numpy 1-d
          arrays.

        - for NxM subplots with N>1 and M>1 are returned as a 2d

```python
#creamos una figura y sus posiciones
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
ax.shape
```

```python
# llamamos a los ejes con la posición matricial correspondiente para pintar en ellos lo que necesitemos
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=12, ha='center')
fig
```

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row');
ax[0,2].plot(np.random.rand(10));
```

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row');
for k,eje in enumerate(ax.ravel()):
    eje.plot(np.random.rand(10))
    eje.set_title('gráfica'+str(k))
```

### Más ejemplos

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title');
```

### Usar `gridspec` para configuraciones más explícitas:

Se pueden agregar celdas creadas generand subfiguras que ocupan varias de ellas

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

```python
fig = plt.figure(figsize=(6, 6))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)

main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)
```

```python
#creamos datos para pintar un histogrma cn densidad
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T
```

```python
# grafo de dispersión en la subfigura más grande
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)
fig
```

```python
# histogramas en las subfiguras izquierda y derecha
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis() #invertimos los ejes para que pegue con la gráfica principal

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
fig
```
