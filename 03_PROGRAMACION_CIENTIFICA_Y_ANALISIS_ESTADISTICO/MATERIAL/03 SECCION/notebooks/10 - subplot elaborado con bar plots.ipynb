# Elaboramos un gr√°fico de barras con ajustes varios

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dat=pd.read_csv('daf.csv',index_col='clus')

dat
```

```python
#creamos el objeto con eje X compartido estableciendo dos filas 5 columnas y el tama√±o de imagen
f,ax=plt.subplots(2,5,sharex=True,figsize=(15,8))

#ajustamos el espacio entre los subplots
f.subplots_adjust(hspace=0.5,wspace=0.5)
#establecemos el supert√≠tulo
f.suptitle('Distribuci√≥n de zonas por cl√∫sters',fontweight='bold',fontsize=30)

#pintamos cada barplot horizontal estableciendo los ticks y los diferentes colores. Ajustamos al centro
for i,k in enumerate(ax.ravel()):
    k.barh(np.arange(12),dat.ix[i+1],height=1,
           tick_label=dat.columns,align='center',color='rrrbbbcccggg')
    k.axis('tight')    
#     k.set_yticklabels(matrizzonas.columns,align)
    
    k.set_xlim(0,23)
#     k.set_xticklabels()
    if i>=5:
        k.set_xlabel('n¬∫ de a√±os de conducci√≥n',)
    k.invert_yaxis()

ejes=ax.ravel()
ejes[0].set_title('Cl 1: Reparaciones varias',fontweight='bold')
ejes[1].set_title('Cl 2: Delantera derecha leve',fontweight='bold')
ejes[2].set_title('Cl 3: Trasera izquierda leve',fontweight='bold')
ejes[3].set_title('Cl 4: Lateral izquierdo en medio',fontweight='bold')
ejes[4].set_title('Cl 5: Delantera izquierda',fontweight='bold')
ejes[5].set_title('Cl 6: Delantera central media altura',fontweight='bold')
ejes[6].set_title('Cl 7: Trasera central media altura',fontweight='bold')
ejes[7].set_title('Cl 8: Delantera derecha grave',fontweight='bold')
ejes[8].set_title('Cl 9: Delantera y trasera abundante',fontweight='bold')
ejes[9].set_title('Cl 10: Delantera hasta habit√°culo',fontweight='bold')

f.tight_layout()
f.subplots_adjust(top=0.88)   
```

```python
#guardamos la figura
f.savefig('figuraguardada.png')
```
