# Regiones y barras de error

Al realizar gráficos se pueden sombrear zonas cuyo significado sea el margen de error en los valores numéricos representados.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.7 * np.random.randn(50)
y = np.sin(x) + dy

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',ecolor='lightgray', elinewidth=3, capsize=0);
```

Con el método `fill_between` se puede sombrear la región que ocupa dos curvas. Esto es adecuado para representar el error en el caso contínuo.

```python
xdata=np.linspace(0,10,10)
ydata=xdata*np.sin(xdata)
plt.plot(xdata,ydata,'or');
```

```python
xdataplus=np.linspace(0,10,1000)
ydataplus=xdataplus*np.sin(xdataplus)

plt.plot(xdata,ydata,'or')
plt.plot(xdataplus,ydataplus);
```

```python
#suponemos que el error se distribuye como sigue
error=xdataplus*np.cos(10*xdataplus)**2
```

```python
# Visualize the result
plt.plot(xdata, ydata, 'or')
plt.plot(xdataplus, ydataplus, '-', color='gray')

plt.plot(xdata,ydata,'or')
plt.plot(xdataplus,ydataplus, '-', color='gray');

plt.fill_between(xdataplus, ydataplus - error, ydataplus + error,
                 color='blue', alpha=0.2)
plt.xlim(0, 10);
plt.title('Error de aproximación de sinusoidal');
```
