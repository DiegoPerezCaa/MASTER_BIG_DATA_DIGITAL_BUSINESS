# Gráficos de densidad y contorno

Son representaciones de la acumulación de elementos a lo largo de la región donde toman valores. Los contornos son líneas de nivel que marcan los puntos a la misma altura.

Las curvas de nivel, particularmente, sirven como *mapa* para visualizar funciones de 3 dimensiones en gráficos con 2 dimensiones.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('bmh')
import numpy as np
```

```python
#esto es una función de dos dimensiones
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
f=lambda x,y:np.sin(x)**10+np.cos(10 + y * x) * np.cos(x) #definición alternativa
```

```python
f(2,3)
```

Numpy incorpora una función que fusiona dos arrays de valores en un único array de dos dimensiones con los puntos formados por las dos listas de coordenadas. Esta función es `meshgrid`

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

La función `plt.contour` genera un mapa de curvas de nivel

```python
# ?plt.contour
```

```python
plt.contour(X, Y, Z, colors='black');
```

El parámetro `cmap` crea un coloreado predefinido tomado de `matplotlib.colors.Colormap`. Información con paletas gráficas en http://matplotlib.org/users/colormaps.html#classes-of-colormaps

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

Si incluímos la función `colorbar` generará una columna representando los colores y los valores numéricos

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

Este otro modo de generar el gráfico de densidades genera un gráfico más suave ó contínuo.

```python
?plt.imshow
```

```python
plt.style.use('default')
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```
