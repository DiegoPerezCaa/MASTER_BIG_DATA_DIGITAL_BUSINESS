# Ajuste de leyendas, colores, textos y anotaciones

### Ajuste de leyendas

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

Se puede usar  `subplots` para pintar varios elementos en la misma figura

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Seno')
ax.plot(x, np.cos(x), '--r', label='Coseno')
ax.axis('equal')
ax.set_title('Funciones trigonom√©tricas')
leg = ax.legend();
```

```python
#frameon sirve para establecer un cuadrado rodeando la leyenda, es True por defecto
ax.legend(loc='lower left', frameon=False,fontsize='x-large',title='Funciones trig.') 
fig
```

```python
#establecemos en dos columnas las leyendas
ax.legend(frameon=False, loc='lower center', ncol=2)
fig
```

- `fancybox` crea una caja de leyenda m√°s estilizada.
- `shadow` agrega sombras a las letras.
- `borderpad` redondea bordes.

```python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
```

```python
x.shape,x[:, np.newaxis].shape, np.arange(0, 2, 0.5).shape
```

El broadcasting de la suma de una matriz 1000x1 con un vector de dimensi√≥n 4 nos crea 4 vectores de dimensi√≥n mil con cada uno de los valores del vector de 4 posiciones sumado.

```python
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
print(y.shape)
lines = plt.plot(x, y,linewidth=4)

# ponemos leyenda s√≥lo para las dos primeras l√≠neas pintadas arriba y abajo izquierda las otras dos
plt.legend(lines[:2], ['primera','segunda']);
```

### Ajuste de colores

```python
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])

plt.imshow(I)
plt.colorbar();
```

```python
plt.imshow(I, cmap='gray');
plt.colorbar();
```

```python
plt.imshow(I, cmap='viridis');
plt.colorbar(orientation='horizontal');
```

```python
x=np.random.randn(150)*100
y=np.random.poisson(100,150)
tipo=np.concatenate((np.repeat(1,50),np.repeat(2,50),np.repeat(3,50)))

plt.scatter(x[:50],y[:50],c='red',label='tipo 1',s=70);
plt.scatter(x[50:100],y[50:100],c='b',label='tipo 2',s=70);
plt.scatter(x[100:150],y[100:150],c='orange',label='tipo 3',s=70);
plt.legend();
```

### Textos y anotaciones

Se puede a√±adir texto en una posici√≥n espec√≠fica del gr√°fico con `plt.text`

```python
plt.scatter(x[:50],y[:50],c='red',label='tipo 1',s=70);
plt.scatter(x[50:100],y[50:100],c='b',label='tipo 2',s=70);
plt.scatter(x[100:150],y[100:150],c='orange',label='tipo 3',s=70);
plt.legend();
plt.axis([-320,310,50,150])

plt.text( 0,72, "Independence Day", fontsize=12,ha='left');
plt.annotate('Esto aqu√≠', xy=( -200,120), xytext=(-200, 140),
            arrowprops=dict(facecolor='black', shrink=0.05),fontsize=22,ha='center');
plt.annotate('bonito',xy=( -200,87), xytext=(-220, 60),
             arrowprops=dict(arrowstyle="fancy",facecolor='red'),
             bbox=dict(boxstyle="round", alpha=0.1),ha='right');
```
