# Scatterplots

Representación de nubes de puntos para analizar la relación de dos variables

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-paper')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
plt.plot(x, y, '-ok');
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

# Función `plt.scatter`

Matplotlib incorpora una función específica para hacer gráficos de dispersión/nube

```python
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # mostrar escala de colores
```

```python
import seaborn as sns
datos=sns.load_dataset('iris')

datos.head()
```

```python
datos.species.factorize()
```

```python
plt.scatter(datos.sepal_length, datos.sepal_width, alpha=0.2,
            s=100*datos.petal_length, c=datos.species.factorize()[0], cmap='viridis')
plt.xlabel(datos.columns[0].replace('_',' '))
plt.ylabel(datos.columns[1]);
```
