# Actividad 1: Numpy

```python
import numpy as np
a=np.random.randint(-10,10,35)
a
```

```python
#vemos que actualmente el array tiene dimensión 1
print(a.shape)
#redefinimos el array con la nueva dimensión
a=a.reshape((5,7))
#pintamos la nueva dimensión así como el array
print(a.shape)
a
```

```python
#seleccionamos los números positivos
a[a>0]
```

```python
#nos quedamos con la lista única eliminando duplicados
aa=np.unique(a)
print(aa)
len(aa) #es la cantidad de elementos que quedan después de eliminar duplicados
```

```python
#obtenemos la media
aa.mean()
```

```python
#calculamos la función solicitada
np.cos(np.exp(aa**2+1)+1)/(1+aa**2)
```

```python
#seleccionamos las entradas menores que -1
aa[aa<-1]
```

```python
#damos un array con los índices de las entradas que son menores que -1
print(np.where(a<-1))
#observamos que nos devuelve dos arrays con los valore de posición de fila y columna respectivamente
list(zip(np.where(a<-1)[0],np.where(a<-1)[1]))
#esto es la lista con los tuples de las posiciones de la matriz 5x7 que verifican la condición
```

```python
#calculamos los máximos, mínimos, y qué posiciones ocupan
print(a.max(),a.min()) #son los valores mínimo y máximo
print(np.unravel_index(a.argmax(), a.shape),np.unravel_index(a.argmin(), a.shape)) #son las posiciones donde se alcanzan
```

```python
#calculamos las sumas de los números positivos y de los números negativos
#observamos que son razonablemente parecidas numéricamente
np.sum(a[a>0]),np.sum(a[a<0])
```

```python
#transformamos el array de nuevo a un vector unidimensional
a.ravel()
```

# Actividad 5: Preprocesamiento de datos

```python
#importamos de nuevo el data-set de coches
#importamos el .csv
import pandas as pd
carr=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',
                 na_values='?',header=None)
carr.head(10)
```

```python
#Creamos una lista de nombres para los atributos para integrarla en el DataFrame
names=['symboling', 'normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels',
 'engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders',
 'engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg',
 'price']
carr.columns=names
carr.head()
```

```python
#borrando instancias vemos cómo son los diferentes tamaños
#se borran
print(carr.dropna().shape,carr.shape)
#probamos borrando instancias con mvs
carr.dropna().head()
```

```python
#introducimos medias en cada columna
mvs={'normalized-losses':carr['normalized-losses'].mean(),'num-of-doors':carr['num-of-doors'].mode().ix[0],
    'bore':carr['bore'].mean(),'stroke':carr['stroke'].mean(),'horsepower':carr['horsepower'].mean(),
    'peak-rpm':carr['peak-rpm'].mean(),'price':carr['price'].mean()}

#sustituimos los valores faltantes por los establecidos en el diccionario
carr.fillna(mvs,inplace=True).head()
```

```python
#si queremos introducir modas ó medianas cambiamos en el diccionario mvs .median ó .mode
```

```python
#normalizamos con min-max
import sklearn.preprocessing as sp
carrminmax=carr.copy()
for k in carr.columns[carr.dtypes!=object]:
    carrminmax[k]=sp.minmax_scale(carr[k], feature_range=(-1, 1), axis=0, copy=True)
carrminmax.head()
```

```python
#normalizamos con normas L1 y L2
import sklearn.preprocessing as sp
import numpy as np
carrL1=carr.copy()
for k in carr.columns[carr.dtypes!=object]:
    carrL1[k]=sp.normalize(carr[k].reshape(1, -1),norm='l1', axis=1, copy=True).ravel()
carrL1.head()
```

```python
#normalizamos l2
import sklearn.preprocessing as sp
import numpy as np
carrL2=carr.copy()
for k in carr.columns[carr.dtypes!=object]:
    carrL2[k]=sp.normalize(carr[k].reshape(1, -1),norm='l2', axis=1, copy=True).ravel()
carrL2.head()
```

```python
#normalizamos Gauss
import sklearn.preprocessing as sp
import numpy as np
carrGauss=carr.copy()
for k in carr.columns[carr.dtypes!=object]:
    carrGauss[k]=sp.scale(carr[k], axis=0, copy=True).ravel()
carrGauss.head(10)
```

```python
#eliminamos las instancias que son un valor atípico para el atributo bore
q1=carr['bore'].quantile(0.25)
q3=carr['bore'].quantile(0.75)
IQ=q3-q1
limsup=q3+1.5*IQ
liminf=q1-1.5*IQ
print(limsup,liminf)
#observamos que en esta variable no tenemos valores atípicos, porque no hemos eliminado ninguna selección
carr.loc[(carr.bore>liminf)&(carr.bore<limsup)].shape, carr.shape
```

```python
import seaborn as sns
%matplotlib inline
sns.pairplot(carr);
```

```python
#creamos variables dummy para los atributos categóricos marcados en el enunciado
pd.get_dummies(carr,columns=['make','fuel-system']).head()
```

# Actividad 6: Análisis Estadistico

```python
#importamos el .csv
import pandas as pd
dat=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data',header=None)
dat.head()
```

```python
#ponemos los nombres a las columnas basándonos en la web citada
dat.columns=['sex','length','diameter','height','whole wheight','schucked weight',
             'viscera weight','shell weight','rings']
dat.head()
```

```python
RR=dat.corr()**2
RR
```

```python
RR.unstack().drop_duplicates().sort_values(ascending=False)
#en la tabla observamos que shell weight es el atributo con mayor correlación con los anillos, 
#aunque es bastante pobre
```

```python
from scipy import stats
stats.ttest_ind(dat.loc[dat.sex=='M','rings'],dat.loc[dat.sex=='F','rings'],equal_var=False)
#el test arroja como resultado que no se pueden considerar la cantidad de anillas de sexo female y male como iguales
#ya que el p-valor <0.05
```

```python
#vemos las medias de anillas de los tres tipos de sexos
dat.loc[dat.sex=='M','rings'].mean(),dat.loc[dat.sex=='F','rings'].mean(),dat.loc[dat.sex=='I','rings'].mean()
```

```python
#contrastamos si los inftiles tienen o no la misma media que los males
stats.ttest_ind(dat.loc[dat.sex=='M','rings'],dat.loc[dat.sex=='I','rings'],equal_var=False)
#el test dice que no, no se pueden considerar la media de sexo infantil como igual a la de males ya que
#el p-valor<0.05
```

```python
#calculamos ahora los coeficientes de la recta de regresión de rings sobre shell weight, que es el atributo
#con el que tiene mayor correlación rings
xx=dat['shell weight']
yy=dat['rings']
#pintamos los datos
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(xx,yy,'o');
```

```python
#calculamos los coeficientes
a=np.cov(xx,yy)[0,1]/np.var(xx)
b=np.mean(yy)-a*np.mean(xx)
#los coeficientes son
a,b
```

```python
#pintamos la recta de regresión sobre la nube de datos
plt.plot(xx,yy,'o',xx,a*xx+b,'r');
```

```python
#la predicción para una concha con shell weight de 0.4 en rings es
str(a*0.4+b)+' rings'
```
