```python
irismv=pd.read_csv('irisMV.dat', comment='@', delimiter=',',header=None,na_values='<null>')
irismv.head()
```

```python
irismv.dropna().iloc[:,4].value_counts() /irismv.iloc[:,4].value_counts()
#número de elementos de cada clase sin MVs, hay una cantidad razonable de cada uno como para hacer introducir MVs
#por categoría
```

```python
#borramos las instancias con MVs
irisborradoinst=irismv.dropna()
irisborradoinst.head()
```

```python
#borramos los atributos con MVs
irisborradoatr=irismv.dropna(axis=1)
irisborradoatr.head()
#este borrado por atributos es absurdo en este caso ya que elimina todos los atributos que tienen MVs
```

```python
import sklearn.preprocessing as sk
#la función que introduce los valores es Imputer en el módulo sklearn.preprocessing
#aquí definimos la función y la estrategia, en este caso la media general del atributo:
imp = sk.Imputer(missing_values='NaN', strategy='mean', axis=0)
irismedia=pd.DataFrame(imp.fit_transform(irismv.iloc[:,0:4],y=irismv.iloc[:,0:4]))
irismedia.head()
```

```python
#introducimos medianas
imp = sk.Imputer(missing_values='NaN', strategy='median', axis=0)
irismedia=pd.DataFrame(imp.fit_transform(irismv.iloc[:,0:4],y=irismv.iloc[:,0:4]))
irismedia.head()
```

```python
#introducimos modas
imp = sk.Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
irismedia=pd.DataFrame(imp.fit_transform(irismv.iloc[:,0:4],y=irismv.iloc[:,0:4]))
irismedia.head()
```

```python
#tomamos el dataframe iris y lo vamos a discretizar
Xdiscret=X.copy()
from sklearn.preprocessing import label_binarize
#creamos los bins (categorías ó intervalos en los que agrupamos)
#en este caso son tres para cada variable, después los montamos en un DataFrame
labels={}
df=pd.DataFrame()
for k in range(0,4):
    labels[k]=['Bajo-'+str(k),'Medio-'+str(k),'Alto-'+str(k)] 
    bins=pd.cut(Xdiscret[k], bins=3,labels=labels[k],retbins=True)[0]
    df=pd.concat([df,pd.DataFrame(label_binarize(bins, classes=labels[k]),columns=labels[k])],axis=1)
    
df.head()
```

# Analizando con estadística

```python
#creamos las variables que vamos a estudiar
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
x=np.random.poisson(16,200)
y=x+np.random.normal(0,5, 200)
plt.plot(x,y,'x');
```

```python
#calculamos los parámetros de la recta de regresión y la pintamos sobre los datos
a1=np.cov(x,y)[0,1]/np.var(x)
b1=np.mean(y)-a1*np.mean(x)
#los coeficientes son
a1,b1
```

```python
#pintamos la recta de regresión sobre la nube de datos
plt.plot(x,y,'x',x,a1*x+b1,'r');
```

```python
#el valor de correlación es
R1=np.cov(x,y)[0,1]/np.sqrt(np.var(x)*np.var(y))
R1
```

```python
x2=np.random.poisson(16,200)
y2=35-x2+np.random.normal(0,1, 200)
plt.plot(x2,y2,'x');
```

```python
#calculamos los parámetros de la recta de regresión y la pintamos sobre los datos
a2=np.cov(x2,y2)[0,1]/np.var(x2)
b2=np.mean(y2)-a2*np.mean(x2)
#los coeficientes son
a2,b2
```

```python
#pintamos la recta de regresión sobre la nube de datos
plt.plot(x2,y2,'x',x2,a2*x2+b2,'r');
```

```python
#el valor de correlación es
R2=np.cov(x2,y2)[0,1]/np.sqrt(np.var(x2)*np.var(y2))
R2
```
